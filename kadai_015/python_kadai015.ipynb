```python
class Human:

# nameとageの引数
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def set_name(self, name):
    self.name = name
  def set_age(self, age):
    self.age = age

  def printinfo(self):
    print('私の名前は',self.name,'です')
    print('私は',self.age,'才です')

#インスタンス化する
human = Human("加藤", 25)

# 値を出力
human.printinfo()
```
