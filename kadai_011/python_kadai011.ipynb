```python
#forの場合
planet_list = ["水", "金", "地", "火", "木", "土", "天", "海", "冥"]
for planet in planet_list:
   print(planet)

#whileの場合1
planet_list = ["水", "金", "地", "火", "木", "土", "天", "海", "冥"]
count = 0
while count < 1:
  for planet in planet_list:
    print(planet)
  count += 1
```

```python
#whileの場合2
#自分で調べた結果このような書き方もありました。
planet_list = ["水", "金", "地", "火", "木", "土", "天", "海", "冥"]
while planet_list:
  print(planet_list.pop(0))
```
