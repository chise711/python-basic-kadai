```python
#ランダムな整数を利用するために、randomモジュールをインポート
import random

# 変数varに0～100までのランダムな整数を代入する
var = random.randint(0, 100)

# 変数varの値を出力
print(var)

# 変数varの値が3の倍数なら「Fizz」、5の倍数なら「Buzz」、3の倍数と5の倍数の両方に該当するなら「FizzBuzz」、どの場合にも該当しない場合「変数varの値」を出力する
if var % 3 ==0 and var % 5 ==0:
    print("FizzBuzz")
elif var % 3 ==0:
    print("Fizz")
elif var % 5 ==0:
    print("Buzz")
else:
    print(var)
```
